## Resolução das Inconsistências

```python
#Resolução da inconsistência encontrada na coluna "QuadroNome". Foi feita a substituição com o replace, dos itens espaçados da colua. 
ES_completo["QuadroNome"].replace(["ACT - Temporário              ", "Aposentadoria                 ", "Bolsista                      ", "CLT                           ", "Cedidos                       ", "Estagiario                    ", "Estatutário-RGPS              ", "Estatutário-RPPS              ", "Inat/Pens Moléstia Grave      ", "Pensao                        ","Residente                     "], ["ACT - Temporário", "Aposentadoria", "Bolsista", "CLT", "Cedidos", "Estagiario", "Estatuário-RGPS", "Estatuário-RPPS", "Inat/Pens Moléstia Grave", "Pensao", "Residente"], inplace=True)
```

```python
#Visualização do resultado após a mudança.
sorted(pd.unique(ES_completo["QuadroNome"]))
```

```python
#Resolução da inconsistência da coluna PlanoCargos que apresentava números em strig como cargo. Relizei a relação de condição, depois de estudar o dataset e fiz a modificação, sanando metade da inconsitencia.
ES_completo.loc[ES_completo.QuadroNome == "Aposentadoria", ['PlanoCargos']] = 'Aposentado'
ES_completo.loc[ES_completo.QuadroNome == "Pensao", ['PlanoCargos']] = 'Pensionista',
ES_completo.loc[ES_completo.QuadroNome == "Inat/Pens Moléstia Grave", ['PlanoCargos']] = 'Inativo/pensionista'
```

```python
#Visualização do resultado após a mudança. Os números com 0 na frente, já foram susbtituidos.
sorted(pd.unique(ES_completo["PlanoCargos"]))
```
